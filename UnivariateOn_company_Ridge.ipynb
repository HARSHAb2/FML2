```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.metrics import mean_absolute_error,r2_score,mean_squared_error
```

```python
data=pd.read_csv("Company_data.csv")
data.head()
```

```python
x = data["TV"].values
y = data["Sales"].values
```

```python
#checking data types of x and y
type(x),type(y)
```

```python
print(x.shape,y.shape)
```

```python
x=x.reshape(-1,1)
y=y.reshape(-1,1)
```

```python
print(x.shape,y.shape)
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y, test_y=train_test_split(x,y,test_size=0.3,random_state=0) 
```

```python
from sklearn import linear_model
z=linear_model.Ridge()
z.fit(train_x,train_y)
```

```python
y_pred = z.predict(test_x)

print(f'MAE = {mean_absolute_error(test_y, y_pred)}')
print(f'MSE = {mean_squared_error(test_y, y_pred)}')
print(f'RMSE = {np.sqrt(mean_squared_error(test_y, y_pred))}')
print(f'R_2 = {r2_score(test_y, y_pred)}')
```
